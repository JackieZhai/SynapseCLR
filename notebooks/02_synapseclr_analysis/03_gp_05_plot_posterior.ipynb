```python
import os
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
from matplotlib import colors

from sklearn.decomposition import PCA

from synapse_utils import io
from synapse_utils.commons import load_imputed_annotations, log1p_zscore

SMALL_SIZE = 10
MEDIUM_SIZE = 12
BIGGER_SIZE = 14

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
repo_root = '../..'
checkpoint_path = '../../output/checkpoint__synapseclr__so3__second_stage'
dataset_path = '../../data/MICrONS__L23__8_8_40__processed'

# these are pre-computed (can be set to None)
contamination_indices_path = os.path.join(
    checkpoint_path, 'indices', 'contamination_meta_df_row_indices.npy')
reload_epoch = 99
node_idx_list = [0, 1, 2, 3]
```

```python
imputed_cell_types_df_path = '../../output/consensus/consensus__meta__uncensored.csv'
imputed_meta_ext_df_path = '../../output/gp/synapse_simclr_production/imputed_meta__rbf__synapse_simclr__ensemble.csv'
```

## 2D Embedding

```python
# # load features
load_prefix = 'encoder.fc'
save_prefix = load_prefix.replace('.', '_')
l2_normalize = False

features_nf, meta_df, meta_ext_df = io.load_features(
    checkpoint_path,
    node_idx_list,
    reload_epoch,
    feature_hook=load_prefix,
    dataset_path=dataset_path,
    l2_normalize=l2_normalize,
    contamination_indices_path=contamination_indices_path)

n_pca_components = 128
if n_pca_components < features_nf.shape[-1]:
    features_nf = PCA(n_components=n_pca_components).fit_transform(features_nf)
```

```python
# helper function
def get_row_indices_from_synapse_ids(synapse_ids):
    synapse_id_to_meta_row_idx_map = {
        synapse_id: row_idx for row_idx, synapse_id in enumerate(meta_df['synapse_id'].values)}
    return [x for x in map(synapse_id_to_meta_row_idx_map.get, synapse_ids)
            if x is not None]
```

## Overlay annotations

```python
# which annotations to use?
annotations_to_use = 'imputed'

if annotations_to_use == 'manual':
    analysis_meta_ext_df = meta_ext_df.copy()
    analysis_meta_ext_df['pre_synaptic_volume_log1p_zscore'] = log1p_zscore(meta_ext_df['pre_synaptic_volume'].values)
    analysis_meta_ext_df['post_synaptic_volume_log1p_zscore'] = log1p_zscore(meta_ext_df['post_synaptic_volume'].values)
    output_subdir = 'manual/'
    alpha = 0.8
    s = 0.5

elif annotations_to_use == 'imputed':
    analysis_meta_ext_df = load_imputed_annotations(
    meta_df,
    imputed_cell_types_df_path,
    imputed_meta_ext_df_path)
    output_subdir = ''
    alpha = 0.5
    s = 0.05
```

```python
# load imputations
imputed_meta_df = pd.read_csv(imputed_meta_ext_df_path, index_col=0)

# load annotation transformation for inverse transformation to original units
loc_and_scale_df = pd.read_csv(
    os.path.join('../../tables/meta_ext_loc_scale.csv'), index_col=None)
```

```python
trait_key_list = [
    'cleft_size_log1p_zscore',
    'presyn_soma_dist_log1p_zscore',
    'postsyn_soma_dist_log1p_zscore',
    'mito_size_pre_vx_log1p_zscore_zi',
    'mito_size_post_vx_log1p_zscore_zi',
    'pre_and_post_cell_types',
    'pre_cell_type',
    'post_cell_type',
    'has_mito_pre',
    'has_mito_post'
]

trait_type_list = [
    'continuous',
    'continuous',
    'continuous',
    'continuous',
    'continuous',
    'categorical',
    'categorical',
    'categorical',
    'categorical',
    'categorical'
]

trait_num_categories_list = [
    None,
    None,
    None,
    None,
    None,
    4,
    2,
    2,
    2,
    2
]

trait_control_list = [
    None,
    None,
    None,
    'has_mito_pre',
    'has_mito_post',
    None,
    None,
    None,
    None,
    None
]

category_labels_list = [
    None,
    None,
    None,
    None,
    None,
    None,
    ['Inhibitory', 'Excitatory'],
    ['Inhibitory', 'Excitatory'],
    ['No', 'Yes'],
    ['No', 'Yes'],
]

category_short_labels_list = [
    None,
    None,
    None,
    None,
    None,
    None,
    ['In', 'Ex'],
    ['In', 'Ex'],
    ['N', 'Y'],
    ['N', 'Y'],
]

category_color_list = [
    None,
    None,
    None,
    None,
    None,
    None,
    ['royalblue', 'deeppink'],
    ['royalblue', 'deeppink'],
    [(0.8, 0.8, 0.8), 'green'],
    [(0.8, 0.8, 0.8), 'royalblue'],
]

trait_scale_list = [
    64e-8,
    1.,
    1.,
    64e-8,
    64e-8,
    1.,
    1.,
    1.,
    1.,
    1.
]
```

```python
for trait_idx in range(len(trait_key_list)):

    trait_key = trait_key_list[trait_idx]
    trait_type = trait_type_list[trait_idx]
    trait_num_categories = trait_num_categories_list[trait_idx]
    trait_control = trait_control_list[trait_idx]
    trait_scale = trait_scale_list[trait_idx]
    
    markercolor = colors.to_rgba('royalblue')[:3] + (0.8,)

    _map = {synapse_id: idx for idx, synapse_id in enumerate(imputed_meta_df['synapse_id'].values)}
    validation_indices = list(map(_map.get, meta_ext_df['synapse_id'].values))

    if trait_type == 'continuous':

        actual_values = meta_ext_df[trait_key].values
        imputed_mean = imputed_meta_df.iloc[validation_indices][f'imputed__{trait_key}__mean'].values
        imputed_std = imputed_meta_df.iloc[validation_indices][f'imputed__{trait_key}__std'].values

        if trait_control is not None:
            keep = meta_ext_df[trait_control].values == True
            actual_values = actual_values[keep]
            imputed_mean = imputed_mean[keep]
            imputed_std = imputed_std[keep]

        loc = loc_and_scale_df[trait_key + '__loc'].item()
        scale = loc_and_scale_df[trait_key + '__scale'].item()

        actual_values_unscaled = np.exp(actual_values * scale + loc) - 1.
        imputed_lo_unscaled = np.exp((imputed_mean - imputed_std) * scale + loc) - 1.
        imputed_mid_unscaled = np.exp((imputed_mean) * scale + loc) - 1.
        imputed_hi_unscaled = np.exp((imputed_mean + imputed_std) * scale + loc) - 1.

        actual_values_unscaled *= trait_scale
        imputed_lo_unscaled *= trait_scale
        imputed_mid_unscaled *= trait_scale
        imputed_hi_unscaled *= trait_scale
        
        fig, ax = plt.subplots(figsize=(2.5, 2.5))

        ax.errorbar(
            actual_values_unscaled,
            imputed_mid_unscaled,
            yerr=np.concatenate((imputed_mid_unscaled - imputed_lo_unscaled[None], imputed_hi_unscaled[None] - imputed_mid_unscaled)),
            fmt='.',
            markersize=2,
            markerfacecolor=markercolor,
            markeredgecolor='none',
            elinewidth=0.5,
            ecolor=(0.1, 0.1, 0.1, 0.5)
        )

        ax.set_xscale('log')
        ax.set_yscale('log')
        ax.set_box_aspect(1.)

        lo = 0.5 * np.sort(actual_values_unscaled)[int(0.01 * len(actual_values_unscaled))]
        hi = 1.5 * np.sort(actual_values_unscaled)[int(0.999 * len(actual_values_unscaled))]
        ax.set_xlim((lo, hi))
        ax.set_ylim((lo, hi))
        ax.plot((lo, hi), (lo, hi), color='gray')
    
        fig.savefig(f'../../output/analysis/imputation/{trait_key}.pdf', bbox_inches='tight')
```

```python
from matplotlib import colors
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
import seaborn as sns

for trait_idx in range(len(trait_key_list)):

    trait_key = trait_key_list[trait_idx]
    trait_type = trait_type_list[trait_idx]
    trait_num_categories = trait_num_categories_list[trait_idx]
    trait_control = trait_control_list[trait_idx]
    short_labels = category_short_labels_list[trait_idx]
    labels = category_labels_list[trait_idx]
    label_colors = category_color_list[trait_idx]
    markercolor = colors.to_rgba('royalblue')[:3] + (0.8,)

    _map = {synapse_id: idx for idx, synapse_id in enumerate(imputed_meta_df['synapse_id'].values)}
    validation_indices = list(map(_map.get, meta_ext_df['synapse_id'].values))

    if trait_type == 'categorical' and trait_num_categories == 2:

        actual_values = meta_ext_df[trait_key].values
        imputed_class_0 = imputed_meta_df.iloc[validation_indices][f'imputed__{trait_key}__class_0'].values
        imputed_class_1 = imputed_meta_df.iloc[validation_indices][f'imputed__{trait_key}__class_1'].values

        sns.reset_orig()
        fig, ax = plt.subplots(figsize=(2.5, 2.5))
        
        ax.hist(imputed_class_0[actual_values == 1], color=label_colors[0], range=(0, 1), bins=20, alpha=0.8, label=labels[0])
        ax.hist(imputed_class_0[actual_values == 0], color=label_colors[1], range=(0, 1), bins=20, alpha=0.8, label=labels[1])
        ax.legend(loc='upper center', fontsize=10)
        
        axins = inset_axes(
            ax, width="100%", height="100%", loc='center',
            bbox_to_anchor=(0.5 - 0.35/2, 0.225, 0.35, 0.35),
            bbox_transform=ax.transAxes)
        
        # calculate confusion matrix
        imputed_values = (imputed_class_1 > 0.5).astype(int)
        c_00 = np.sum((actual_values == 0) & (imputed_values == 0))
        c_01 = np.sum((actual_values == 0) & (imputed_values == 1))
        c_10 = np.sum((actual_values == 1) & (imputed_values == 0))
        c_11 = np.sum((actual_values == 1) & (imputed_values == 1))
        p_00 = c_00 / (c_00 + c_01)
        p_01 = c_01 / (c_00 + c_01)
        p_10 = c_10 / (c_10 + c_11)
        p_11 = c_11 / (c_10 + c_11)
        cmat = np.asarray([[p_11, p_10], [p_01, p_00]])
        sns.set_context("notebook", font_scale=0.7)
        res = sns.heatmap(cmat, annot=True, ax=axins, cbar=False, cmap=plt.cm.Blues, fmt='.2f',
                          xticklabels=short_labels, yticklabels=short_labels)
        for _, spine in res.spines.items():
            spine.set_visible(True)
        axins.yaxis.set_label_position("right")
        axins.set_ylabel('Actual', fontsize=10)
        axins.set_title('Predicted', fontsize=10, pad=0)
        
        fig.savefig(f'../../output/analysis/imputation/{trait_key}.pdf', bbox_inches='tight')
```
